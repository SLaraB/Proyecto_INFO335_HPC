```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.preprocessing import LabelEncoder


densenet = pd.read_csv("densenet_results.csv")
inception = pd.read_csv("inception_results.csv")
resnet = pd.read_csv("resnet_results.csv")
vgg16 = pd.read_csv("vgg16_results.csv")
vgg19 = pd.read_csv("vgg19_results.csv")


```

```python
# Asegúrate de que la columna 'image_figures' es del tipo correcto en todos los dataframes
densenet['image_figures'] = densenet['image_figures'].astype(str)
inception['image_figures'] = inception['image_figures'].astype(str)
resnet['image_figures'] = resnet['image_figures'].astype(str)
vgg16['image_figures'] = vgg16['image_figures'].astype(str)
vgg19['image_figures'] = vgg19['image_figures'].astype(str)
```

```python
# Renombra las columnas de 'labels' para que sean únicas
densenet = densenet.rename(columns={'labels': 'labels_densenet'})
inception = inception.rename(columns={'labels': 'labels_inception'})
resnet = resnet.rename(columns={'labels': 'labels_resnet'})
vgg16 = vgg16.rename(columns={'labels': 'labels_vgg16'})
vgg19 = vgg19.rename(columns={'labels': 'labels_vgg19'})
```

```python
# Une los dataframes en uno solo basado en la columna 'image_figures'
df = pd.merge(densenet, inception, on='image_figures', how='outer')
df = pd.merge(df, resnet, on='image_figures', how='outer')
df = pd.merge(df, vgg16, on='image_figures', how='outer')
df = pd.merge(df, vgg19, on='image_figures', how='outer')
```
